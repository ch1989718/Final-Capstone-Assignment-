# Chinese Restaurant in Toronto or New York?


## Inntroduction

Due to the developing technology, people are getting closer. You can find other countries food in your country.     
People travel all over the world to try different culture and food. Some people will look for an opportunity     
to start a food business in another country. However, they need to do some researches to decide which city     
is a good location to start their future.

## Data

The data will be found based on the Fopursquare server. The data will include top 50 restaurants in Toronto and New    
York. These JSON information will include the locations, type of food, what country food. Based on these data and     
decide which place will have more profits to start a new restaurant.

## Methodology

The number of restaurants will be collected in both Toronto and New York. The population is the most important factor     
of the restaurant. These date will be transformed into a data frame and visualized in a map.

## Results

Setup the radius =1000, New York has 29 Chinese restaurants. Toronto has 21 restaurants. New York has more    
population than Toronto and the restaurants in New Yokr are more concentrated than in Toronto.

## Discussion

New York has more population and concentrated. Choosing to open a Chinese restaurant in New York will face a       
higher competition and higher budget, but more population is a big potential market. Choosing Toronto will have     
a less competition and less budget. The less population will be a problem for a long term.

## Conclusion

To conclude, New York will be a better place to open a restaurant. Even though under a higher competition and     
budget, the higher population will be the best reason to overcome these issues. Besides, New York has more tourist    
spots to attract more people from all of the world and that is a potential profit as well.

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
search_query = 'Chinese'
radius = 1000
print(search_query + ' .... OK!')
url='https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID,CLIENT_SECRET,latitude,longitude,VERSION,search_query,radius,LIMIT)
url
results = requests.get(url).json()
results
venues = results['response']['venues']

dataframe = json_normalize(venues)
dataframe.head()
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
dataframe_filtered.name
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=15) 

folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='Conrad Hotel',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

for lat, lng, label in zip(dataframe_filtered.lat, dataframe_filtered.lng, dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)
venues_map
```
